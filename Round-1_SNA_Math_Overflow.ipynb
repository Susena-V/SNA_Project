```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# File path
file_path = "/content/sx-mathoverflow.txt"

# Load edges from file, ignoring comments
with open(file_path, "r") as f:
    edges = [line.strip().split() for line in f if not line.startswith("#")]

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from((int(u), int(v)) for u, v, *_ in edges)  # Ignores extra temporal columns

# Compute degree statistics
in_degrees = dict(G.in_degree())
out_degrees = dict(G.out_degree())
degrees = dict(G.degree())  # Normal degree (sum of in and out for directed graphs)

# Convert to lists for plotting
in_degree_values = list(in_degrees.values())
out_degree_values = list(out_degrees.values())
degree_values = list(degrees.values())

# Compute statistics
degree_stats = {
    "In-Degree": (max(in_degree_values), min(in_degree_values), np.mean(in_degree_values), np.std(in_degree_values)),
    "Out-Degree": (max(out_degree_values), min(out_degree_values), np.mean(out_degree_values), np.std(out_degree_values)),
    "Total Degree": (max(degree_values), min(degree_values), np.mean(degree_values), np.std(degree_values)),
}

# Print statistics
for degree_type, (max_val, min_val, avg, std) in degree_stats.items():
    print(f"{degree_type}: Max: {max_val}, Min: {min_val}, Avg: {avg:.2f}, Std Dev: {std:.2f}")

# Plot distributions
plt.figure(figsize=(12, 5))

plt.hist(in_degree_values, bins=30, color='green', alpha=0.6, label='In-Degree')
plt.hist(out_degree_values, bins=30, color='orange', alpha=0.6, label='Out-Degree')
plt.hist(degree_values, bins=30, color='blue', alpha=0.5, label='Total Degree')

plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree, In-Degree, and Out-Degree Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import networkx as nx

# Compute centralities
in_degree_centrality = nx.in_degree_centrality(G)
out_degree_centrality = nx.out_degree_centrality(G)
degree_centrality = nx.degree_centrality(G)  # For undirected influence

# Print sample results
print("Sample In-Degree Centrality:", list(in_degree_centrality.items())[:5])
print("Sample Out-Degree Centrality:", list(out_degree_centrality.items())[:5])
print("Sample Degree Centrality:", list(degree_centrality.items())[:5])
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)
```

```python
def compute_alpha(G):
    A = nx.to_numpy_array(G)
    eigenvalues = np.linalg.eigvals(A)
    lambda_max = max(abs(eigenvalues))
    return 0.85 / lambda_max

# Compute Katz centrality with adaptive alpha
alpha = compute_alpha(G)
try:
    katz_centrality = nx.katz_centrality(G, alpha=alpha, beta=1.0, max_iter=300, tol=1e-6)
except nx.PowerIterationFailedConvergence:
    katz_centrality = nx.katz_centrality(G, alpha=alpha/2, beta=1.0, max_iter=300, tol=1e-6)
```
